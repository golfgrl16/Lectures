```python
# can easily loop through lists

import random

def rsum():
  """
      Sum up 100 uniform random numbers
  """ 
  #List comprehension
  rnums=[random.uniform(0,1) for n in range(0,100)]

  mysum=0.0

  for a in rnums:
    #do something
    mysum+=a
  
  return(mysum)

"""
#Now do that a bunch 
all_answers=[]
for i in range(1000):
  all_answers.append(rsum()) 
"""

all_answers=[rsum() for i in range(1000)]
```

```python
print(all_answers)
```

```python
import seaborn as sbn

sbn.distplot(all_answers, rug=True, color="green");
```

Now let's find the number of our random numbers which are greater than 0.9?



```python
challenge_nums=[random.uniform(0,1) for i in range(1000)]

#How many of your numbers are greater than 0.9?
num_count=0
for entry in challenge_nums:
  if entry>0.9:
    num_count+=1
  else:
    num_count-=1
print(num_count)
```

## Map, Filter, zip and Other Tricks

Many times we will want to apply a function to a list on inputs. For this we can use the map function. 

```python
import numpy as np
import pylab as plt

x=np.arange(0,2*np.pi,0.01)

print(x)
```

```python
y=[]

for val in x:
  y.append(np.sin(val))

plt.plot(x,y);
```

```python
y=map(np.sin,x)

plt.plot(x,list(y));
```

```python
pos_y=filter(lambda a: a>=0.0, y)

print(list(pos_y))
```
